### Data cleaning

```python
cleaned_data = data.replace({'%':'', '\$' : '', ',':'', '\(': '' , '\)':''}, regex=True)
cleaned_data.head()
```

```python
cleaned_data.info(0)
```

```python
for index, column in enumerate(cleaned_data.columns):
    if 6 < index < 68:
        cleaned_data[column] = pd.to_numeric(cleaned_data[column])
cleaned_data.info(0)
```

### Data Exploration

```python
sns.distplot(cleaned_data['Price'], kde=False)
```

```python
sns.distplot(cleaned_data[cleaned_data['Price'] < 150]['Price'], kde=True)
```

```python
sectors = cleaned_data.groupby(['Sector']).agg({'Price' : 'mean'})
sectors
```

```python
plot = sns.barplot(sectors.index, sectors['Price'])
_ = plot.set_xticklabels(plot.get_xticklabels(), rotation=90)
```

```python
industries = cleaned_data.groupby(['Industry']).agg({'Price' : 'mean'})
industries
```

```python
industry_sector = cleaned_data[['Industry', 'Sector']]
financial_industry = industry_sector[industry_sector['Sector'] == 'Financial']['Industry'].drop_duplicates().values
```

```python
financial_industries_prices = industries[industries.index.isin(financial_industry)]
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
plot = sns.barplot(financial_industries_prices.index, financial_industries_prices['Price'], ax=ax)
_ = plot.set_xticklabels(plot.get_xticklabels(), rotation=90)
```

```python
prop_and_casaulty = cleaned_data[cleaned_data['Industry'] == 'Property & Casualty Insurance']
prop_and_casaulty.head()
```

```python
fig, ax = plt.subplots(figsize=(20,5))
plot = sns.barplot(prop_and_casaulty['Company'], prop_and_casaulty['Price'], ax=ax)
_ = plot.set_xticklabels(plot.get_xticklabels(), rotation=90)
```

```python
outlier_removed_data = cleaned_data[cleaned_data['Ticker'] != 'BRK-A']
sectors_outlier_removed = outlier_removed_data.groupby(['Sector']).agg({'Price' : 'mean'})
sectors_outlier_removed
```

```python
fig, ax = plt.subplots(figsize=(10,5))
plot = sns.barplot(sectors_outlier_removed.index, sectors_outlier_removed['Price'], ax=ax)
_ = plot.set_xticklabels(plot.get_xticklabels(), rotation=90)
```
